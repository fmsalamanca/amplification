```python
import numpy as np
from scipy.special import erf, erfc

def Ac(Nc,Ne):
    a = (3*Ne)/(2*Nc)
    return 0.5 + (np.sqrt(a/np.pi)*np.exp(-a))/erf(np.sqrt(a))

def func(Nc):
        return Nc - rho*R*T*Ac(Nc,Ne)/(gc*Ms)
```

```python
rho = 0.92 #g/cm3
T   = 298 #K
R   = 8.3144 #J/(mol K)
Ms  = 131 #g/mol
Na  = 6.022*np.float_power(10,23) #Avogadro constant

z = 0.387 #From revisiting segmental order...
```

```python
from scipy.optimize import fsolve

gc = 0.039
ge = 0.130

Ne  = (1/np.sqrt(6))*rho*R*T/(ge*Ms)
Nc  = fsolve(func,x0=2)[0] #Ncc is an array, so we select the component zero which is a float

Nc = 13
Ne = 470
n      = Nc*(1-Ac(Nc,Ne))/(2*Ac(Nc,Ne))
m      = (2/Nc)*z*np.sqrt((Nc+2*n)/(0.8*Ne))*np.arctan(Nc/(2*np.sqrt(Nc*n+n**2)))

print("Nc = ", int(np.round(Nc,0)))
print("Ne = ", int(np.round(Ne,0)))
print("Ac = ", np.round(Ac(Nc,Ne),3))
print("m = ", np.round(m,6))
```
